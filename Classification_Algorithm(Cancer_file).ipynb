# Performance Comparison of ML Models on Breast Cancer Dataset

# Overview Of Problem Statement:

Breast cancer is one of the most common and life-threatening diseases affecting women worldwide. Early and accurate diagnosis plays a crucial role in improving patient outcomes. Machine learning techniques provide an effective way to analyze medical datasets and assist in predicting diagnoses based on various clinical features.

In this project, the Breast Cancer dataset from the sklearn library is utilized to explore the potential of supervised learning algorithms in classifying tumors as malignant or benign. The primary challenge lies in preprocessing the dataset, selecting the most effective classification algorithms, and identifying the best-performing model through comparative evaluation.

# Objective:

To apply supervised learning techniques to classify breast cancer cases using the dataset provided in the sklearn library. This involves evaluating the performance of multiple classification algorithms to determine the most suitable model for this dataset.


# Data Description:

**Source:** The dataset used in this project is the Breast Cancer dataset provided by the sklearn library. This dataset is widely used for benchmarking machine learning algorithms in medical applications.

**Features:** The dataset contains 30 numerical features derived from a digitized image of a breast mass. These features describe the characteristics of cell nuclei present in the image, grouped into the following categories:

   * **Radius (mean, standard error, worst):** Measurement of mean distance from the center to points on the perimeter.
   * **Texture (mean, standard error, worst):** Standard deviation of gray-scale values.
   * **Perimeter (mean, standard error, worst):** Measurement of the tumor perimeter.
   * **Area (mean, standard error, worst):** Size of the tumor area.
   * **Smoothness (mean, standard error, worst):** Degree of variation in the radius lengths.
   * **Compactness (mean, standard error, worst):** Combination of perimeter and area.
   * **Concavity (mean, standard error, worst):** Severity of concave portions of the contour.
   * **Concave Points (mean, standard error, worst):** Number of concave portions of the contour.
   * **Symmetry (mean, standard error, worst):** Symmetry of the cell nuclei.
   * **Fractal Dimension (mean, standard error, worst):** Metric for measuring complexity or irregularity of the tumor boundary.
   * **Target Variable:** A binary classification target indicating whether the tumor is: Malignant (coded as 1) OR Benign (coded as 0)

**Dataset Size:**
* **Samples:** 569 observations.
* **Features:** 30 numerical features (plus the target variable).

### checking outliers using statistical method(IQR)

```python
# Calculate Q1, Q3, and IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = ((df < lower_bound) | (df > upper_bound))
print("Number of outliers in each column:")
print(outliers.sum())
```

```python
# Cap outliers to the threshold values (IQR method)
df_capped = df.copy()
df_capped = df_capped.clip(lower=lower_bound, upper=upper_bound, axis=1)

print("Shape after capping outliers:", df_capped.shape)
```

## Split Data into Training and Testing Sets

## Build the ML Model:

### Classification Algorithm Implementation

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix #Evaluation Metrics

from sklearn.feature_selection import SelectKBest, f_classif
```

```python
# Scale the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

**1. LogisticRegression:**

Logistic Regression is a linear model that uses a logistic function to estimate probabilities of binary outcomes (0 or 1).
Applications: Email spam detection, disease diagnosis, credit scoring.
Suitability:It works well for binary classification problems, especially when there is a linear relationship between the features and the target variable. It is easy to interpret, which is helpful in medical datasets like breast cancer data, where understanding the influence of each feature on the classification result can be crucial.

**2. DecisionTreeClassifier:**

Decision Trees split the data into subsets based on feature values, creating a tree-like structure where each node represents a decision on a feature, and each leaf node represents a class label.
Applications: Risk assessment, fraud detection, customer segmentation.
Suitability: Decision Trees can capture non-linear relationships between features and the target. They are also easy to interpret and visualize, making them useful for explaining the classification decisions. This is beneficial for medical datasets where interpretability is valuable, although they can be prone to overfitting.

**3. KNeighborsClassifier:**

k-NN is a non-parametric algorithm that classifies data points based on the labels of the k nearest data points in the feature space. It is highly dependent on the choice of distance metric and the value of k. No explicit training phase, making it a lazy learner. Applications: Recommender systems, pattern recognition, anomaly detection.
Suitability: k-NN is simple and can capture non-linear relationships, but it can be sensitive to noisy features and may struggle with high-dimensional data if the dataset has many features. It might be less suitable for large datasets as it requires storing and comparing all points, which can be computationally expensive.

**4. Random Forest Classifier:**
 
Random Forest is an ensemble method that builds multiple decision trees on random subsets of the data and aggregates their predictions. This helps to reduce overfitting and improves generalization. Handles large datasets with higher dimensionality. Applications: Financial forecasting, image classification, healthcare diagnostics.
Suitability: Random Forest is robust to noise and handles complex, high-dimensional datasets well. It’s suitable for this dataset as it reduces overfitting (a common issue with Decision Trees) and can provide feature importance, which might help identify which features are most relevant in predicting breast cancer.



**5. Support Vector Machine (SVM):**

SVM finds the hyperplane that maximizes the margin between classes, meaning it looks for the boundary that separates classes with the largest possible distance from any point in either class. Applications: Image classification, text categorization, bioinformatics.
Suitability: SVM is effective in high-dimensional spaces and can handle non-linear boundaries using kernel functions. If the dataset has clear class separability, SVM is likely to perform well. However, SVMs can be slower with large datasets and harder to interpret, which may be a limitation in some applications.

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize a dictionary to store results
results = []

# Models dictionary with fitted classifiers
models = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "k-NN": KNeighborsClassifier()
}

# Evaluate each model
for name, model in models.items():  # {'name':model()} in models
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    y_prob = model.predict_proba(x_test)[:, 1] if hasattr(model, "predict_proba") else model.decision_function(x_test)

    # Calculate metrics
    metrics = {
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "ROC AUC": roc_auc_score(y_test, y_prob) if y_prob is not None else "N/A"
    }
    results.append(metrics)

    # Display metrics for the current model
    #print(f"\n{name} Metrics:")
    for metric, value in metrics.items():
        if value is not None:
            try:
                print(f"{metric}: {value:.4f}")
            except ValueError:
                print(f"{metric}: {value}")  # Print as-is if not a float

    # Plot confusion matrix for each model
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix for {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
```

## Model Comparison

## Compare the performance of the five classification algorithms.

```python
# Display the comparison table
print("\nModel Performance Comparison:")
print(results_df)
# Plot comparison metrics
results_df.set_index("Model").plot(kind='bar', figsize=(10, 6), colormap='coolwarm', rot=45, title="Model Performance Comparison")
plt.ylabel('Score')
plt.xlabel('Model')
plt.legend(title='Metric')
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside
plt.show()
```

### Which algorithm performed the best and which one performed the worst?

```python
# Find the best and worst-performing algorithms
best_model_row = results_df.loc[results_df['Accuracy'].idxmax()]
best_model_name = best_model_row['Model'] 
worst_model_row = results_df.loc[results_df['Accuracy'].idxmin()]

print("\nBest Performing Model:")
print(best_model_row)

print("\nWorst Performing Model:")
print(worst_model_row)
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter tuning for the best-performing model
if isinstance(models[best_model_name], LogisticRegression):
    param_grid = {
        'penalty': ['l1', 'l2', 'elasticnet', 'none'],
        'C': [0.01, 0.1, 1, 10, 100],
        'solver': ['lbfgs', 'saga', 'liblinear'],
        'max_iter': [1000, 2000]
    }

elif isinstance(models[best_model_name], RandomForestClassifier):
    param_grid = {
        'n_estimators': [50, 100, 200],
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'bootstrap': [True, False]
    }

elif isinstance(models[best_model_name], KNeighborsClassifier):
    param_grid = {
        'n_neighbors': [3, 5, 7, 9, 11],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan', 'minkowski']
    }
else:
    raise ValueError(f"No parameter grid defined for {best_model_name}")
```

```python
# Perform GridSearchCV
if param_grid:
    grid_search = GridSearchCV(estimator=models[best_model_name], param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(x_train_scaled, y_train)

# Display the best parameters and the corresponding score
#print("\nBest Parameters for", best_model_name, ":", grid_search.best_params_)
#print("Best Cross-Validated Accuracy:", grid_search.best_score_)

# Optionally refit the best model on the full training set
    best_model_name = grid_search.best_estimator_
    print("Best parameters:", grid_search.best_params_)
    #best_tuned_model.fit(x_train_scaled, y_train)
```
